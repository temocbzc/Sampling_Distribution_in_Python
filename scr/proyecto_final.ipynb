### Programa que realiza todas las muestras posibles de diferentes tamaños, de una población dada.

#### Asignatura: Estadística y Probabilidad 2

El presente programa esta diseñado para ser de ayuda en la Unidad 2: Distribuciones Muestrales de la asignatura de Estadística II. Por lo regular muchuos de los softwares en el mercado, realizan este muestreo bajo el enfoque frecuentista, sin embargo el programa diseñado realiza todas las muestras posibles y no solo hace una aproximación frecuentista.

```python
#Librerias a utilizar

import pandas as pd
import numpy as np
import tkinter as tk
import tkinter.filedialog
import PIL
from tkinter import ttk,messagebox,filedialog
from PIL import ImageTk
import PIL.Image
import os
import matplotlib.pyplot as plt
from matplotlib.backends.backend_tkagg import (FigureCanvasTkAgg,NavigationToolbar2Tk)
from os import remove
from matplotlib.figure import Figure

from tkinter import *
from tkinter import ttk
import matplotlib.pyplot as plt
from IPython.display import clear_output


#Algunas variables globales a utilizar

n=0
e=[]
s=[]
s2=[]
#Funciones a utilizar

#La siguiente función sirve para mostrar los resultados en pantalla

def resultados(result):
    lb4.config(text=result)

#La siguiente función sirve para subir un archivo tipo CSV que contiene la población
    
def cargar_archivo():
    global e
    archivo =tk.filedialog.askopenfilename(title="Abrir Archivo", filetypes=(("Archivos CSV","*.csv"), ("Todos los Archivos","*.*")))
    edades = pd.read_csv(archivo)
    e=edades.to_numpy().tolist()

#Las siguientes funciones sirven para realizar los muestre de tamaño n=2, 3, 4 o 5     
    
def n2(): global n; n=2; 
def n3(): global n; n=3; 
def n4(): global n; n=4; 
def n5(): global n; n=5; 

#Función para graficar medias muestrales

def grafica_medias():
    global frame2
    x=s2.value_counts()   
    frame2.destroy()
    frame2 = tk.Frame(ventana,bd=4,relief=tk.RAISED)
    frame2.place(x=700,y=140,width=840,height=550)
    fig=Figure()
    grafica=fig.add_subplot(111)
    grafica.bar(x.index,x/len(s2),0.1)
    grafica.set_xlabel(r'Medias muestrales $(\bar{x})$')
    grafica.set_ylabel(r'$P(\bar{x})$')  
    canvas = FigureCanvasTkAgg(fig, master = frame2)
    canvas.get_tk_widget().pack(fill=tk.BOTH, expand=1)
    canvas.draw()

#La siguiente función sirve para guardar un archivo CSV   
    
def guardar_archivo():
    global s
    path=filedialog.asksaveasfilename(defaultextension=".csv")
    s.value_counts().to_csv(path)
    messagebox.showwarning("Éxito", 'El archivo se ha guardado')
    
def guardar_archivo2():
    global s2
    path=filedialog.asksaveasfilename(defaultextension=".csv")
    s2.value_counts().to_csv(path)
    messagebox.showwarning("Éxito", 'El archivo se ha guardado')
    
#La siguiente función sirve para realizar las medias muestrales

def medias_muestrales():
    global s
    global s2
    medias=[]
    for i in range(len(s)):
        medias.append(np.mean(s[i]))
    s2=pd.Series(medias)
    texto=s2.value_counts() #.head()
    resultados(texto)
    del medias
    
    btn4 = Button(ventana,text="Guardar Medias", command=guardar_archivo2)
    btn4.place(x=400, y=480)
    
    btn5 = Button(ventana,text="Graficar Medias Muestrales", command=grafica_medias)
    btn5.place(x=520, y=480)
    
def muestrea():
    a=[]
    global s
    if n==2: 
      
        for i in range(len(e)):
            b=e[i][0]
            a.append(b)
        muestras = []
        for i in range(len(e)):
            for j in range(len(e)):
                muestras.append([a[i],a[j]])
        
        s = pd.Series(muestras)
        texto=s.value_counts().head(16)
        resultados(texto)
        
    
    elif n==3:
     
        for i in range(len(e)):
            b=e[i][0]
            a.append(b)
        muestras = []
        for i in range(len(e)):
            for j in range(len(e)):
                for h in range(len(e)):
                    muestras.append([a[i],a[j],a[h]])
        s = pd.Series(muestras)
        texto=s.value_counts()
        resultados(texto)
    
    elif n==4:
        for i in range(len(e)):
            b=e[i][0]
            a.append(b)
        muestras = []
        for i in range(len(e)):
            for j in range(len(e)):
                for h in range(len(e)):
                    for k in range (len(e)):
                        muestras.append([a[i],a[j],a[h],a[k]])
        s = pd.Series(muestras)
        texto=s.value_counts()
        resultados(texto)
    
    elif n==5:
        for i in range(len(e)):
            b=e[i][0]
            a.append(b)
        muestras = []
        for i in range(len(e)):
            for j in range(len(e)):
                for h in range(len(e)):
                    for k in range (len(e)):
                        for m in range(len(e)):
                            muestras.append([a[i],a[j],a[h],a[k],a[m]])
        s = pd.Series(muestras)
        texto=s.value_counts()
        resultados(texto)
        
    else:
        tk.messagebox.showwarning(title="Error", message="No has seleccionado el tamaño de las muestras")

    btn2 = Button(ventana,text="Guardar Muestras", command=guardar_archivo)
    btn2.place(x=400, y=440)
    
    btn3 = Button(ventana,text="Calcular Medias Muestrales", command=medias_muestrales)
    btn3.place(x=520, y=440)
    
#Cuerpo principal del programa

ventana = Tk()
ventana.title("Distribuciones Muestrales")
ventana.geometry("1550x700+0+0")

frame1 = tk.Frame(ventana,bd=4,relief=tk.RAISED)
frame1.place(x=390,y=140,width=300,height=550)

frame2 = tk.Frame(ventana,bd=4,relief=tk.RAISED)
frame2.place(x=700,y=140,width=840,height=550)

x=pd.Series([0,0,0],index=[0,0,0])


lb1=Label(ventana,text="\n\nPrograma que realiza todas las muestras posibles de tamaño n \n de una población de tamaño N con reemplazo",
        font=("arial",18))

lb1.pack()

lb2=Label(ventana,text="\n\n\n            Cargar datos de la Población\n\n\n",font=("arial",14))
lb2.pack(anchor=NW)

btn = Button(ventana,text="Cargar Archivo", command=cargar_archivo)
btn.place(x=140, y=215)
#btn.pack()

lb3=Label(ventana,text="            Elige el tamaño de las muestras\n",font=("arial",14))
lb3.pack(anchor=NW)

lb4=Label(ventana,text=" ",font=("arial",10))
lb4.place(x=430, y=150)

v = tk.IntVar()
tk.Radiobutton(ventana, text="n=2", padx = 20, value=2, command=n2).place(x=140,y=300)
tk.Radiobutton(ventana, text="n=3", padx = 20, value=3, command=n3).place(x=140,y=330)
tk.Radiobutton(ventana, text="n=4", padx = 20, value=4, command=n4).place(x=140,y=360)
tk.Radiobutton(ventana, text="n=5", padx = 20, value=5, command=n5).place(x=140,y=390)

start = Button(ventana,text="         Start         ", command=muestrea)
start.place(x=140, y=420)

ventana.mainloop()
```
